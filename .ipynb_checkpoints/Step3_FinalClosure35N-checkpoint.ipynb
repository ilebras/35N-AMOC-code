Extended, no LW version

```python
LWmerge_ext['Ekman_wdepth'],na=xr.broadcast(CCMP.Ekman_wdist/ekd*10,LWmerge_ext.depth.sel(depth=slice(0,ekd)))
LWmerge_ext['Ekman_wdepth']=LWmerge_ext['Ekman_wdepth'].fillna(0)
```

Add Ekman

```python
LWmerge['trans_wEk']=LWmerge.trans+LWmerge.Ekman_wdepth
LWmerge_ext['trans_wEk']=LWmerge_ext.trans+LWmerge_ext.Ekman_wdepth
```

```python
LWmerge_ext.Ekman_wdepth.sum(dim=['depth','dist']).plot()
LWmerge.Ekman_wdepth.sum(dim=['depth','dist']).plot(ls='--')
```

```python
LWmerge['trans_wEk_extrap']=LWmerge['trans_wEk'].copy()
LWmerge_ext['trans_wEk_extrap']=LWmerge_ext['trans_wEk'].copy()
```

```python
for tt in range(len(LWmerge.date)):
    for dd in range(len(LWmerge.dist)):
        if ~isnan(LWmerge.bathy_i[dd]):
            LWmerge['trans_wEk_extrap'][LWmerge.depth==(int(-LWmerge.bathy_i[dd]/10)*10.0),dd,tt]=0

LWmerge['trans_wEk_extrap']=LWmerge['trans_wEk_extrap'].interpolate_na(dim='depth')
```

```python
LWmerge_ext=LWmerge_ext.interp(depth=arange(0,6000,10))
```

```python
for tt in range(len(LWmerge_ext.date)):
    for dd in range(len(LWmerge.dist)):
        if ~isnan(LWmerge.bathy_i[dd]):
            LWmerge_ext['trans_wEk_extrap'][LWmerge_ext.depth==(int(-LWmerge.bathy_i[dd]/10)*10.0),dd,tt]=0

LWmerge_ext['trans_wEk_extrap']=LWmerge_ext['trans_wEk_extrap'].interpolate_na(dim='depth')
```

#### Add compensation transport

```python
#separation of western boundary area and interior
wbc_dist2=365
```

```python
dist_mat,depth_mat=xr.broadcast(LWmerge['distdiff'],LWmerge['depthdiff'])
LWmerge['area']=LWmerge['distdiff']*LWmerge['depthdiff'].where(LWmerge.depth<-LWmerge.bathy_i)
```

```python
LWmerge['area_deep_east']=LWmerge.area.where(LWmerge.dist>wbc_dist2).where(LWmerge.depth>2000)
LWmerge['area_deep_all']=LWmerge.area.where(LWmerge.depth>2000)
```

```python
throughflow=-1
```

```python
LWmerge['trans_comp']=((-LWmerge.trans_wEk).sum(dim=['dist','depth'])+throughflow)/LWmerge.area_deep_east.sum(dim=['depth','dist'])*LWmerge.area_deep_east
LWmerge['trans_comp_extrap']=((-LWmerge.trans_wEk_extrap).sum(dim=['dist','depth'])+throughflow)/LWmerge.area.sum(dim=['depth','dist'])*LWmerge.area
```

```python
LWmerge_ext['trans_comp']=((-LWmerge_ext.trans_wEk).sum(dim=['dist','depth'])+throughflow)/LWmerge.area_deep_all.sum(dim=['depth','dist'])*LWmerge.area_deep_all
LWmerge_ext['trans_comp_extrap']=((-LWmerge_ext.trans_wEk_extrap).sum(dim=['dist','depth'])+throughflow)/LWmerge.area.sum(dim=['depth','dist'])*LWmerge.area
```

```python
(LWmerge.trans_comp*1e6*100/LWmerge.area_deep_east).mean(dim=['depth','dist']).plot(label='unobs comp')
(LWmerge.trans_comp_extrap*1e6*100/LWmerge.area).mean(dim=['depth','dist']).plot(label='extrap')
(LWmerge_ext.trans_comp*1e6*100/LWmerge.area_deep_all).mean(dim=['depth','dist']).plot(label='no LW-unobs comp ext')
(LWmerge_ext.trans_comp_extrap*1e6*100/LWmerge.area).mean(dim=['depth','dist']).plot(label='no LW-extrap ext')
grid('on')
legend()
ylabel('compensation velocity [cm/s]')
# savefig(figdir+'CompVel.png',dpi=300,bbox_inches='tight')
```

Note: autocorrelation and timescale anaylsis for all time series in Step 4

```python
neff=10*4
neff_ext=15*4
```

```python
tstar=2.021 # 95% confidence for 40 effective degrees of freedom
tstar_ext=2.0 # 95% confidence for 60 effective degrees of freedom
```

```python
conf_int_extrap=(LWmerge.MOCdepth_var_extrap.std(dim='date')*tstar/LWmerge.time.std(dim='date')/sqrt(neff-2)).values
conf_int_unobs=(LWmerge.MOCdepth_var.std(dim='date')*tstar/LWmerge.time.std(dim='date')/sqrt(neff-2)).values
conf_int_noLW_extrap=(LWmerge_ext.MOCdepth_var_extrap.sel(date=slice(date_start,date_end)).std(dim='date')*tstar/LWmerge.time.std(dim='date')/sqrt(neff-2)).values
conf_int_noLW_unobs=(LWmerge_ext.MOCdepth_var.sel(date=slice(date_start,date_end)).std(dim='date')*tstar/LWmerge.time.std(dim='date')/sqrt(neff-2)).values
conf_int_ext_extrap=(LWmerge_ext.MOCdepth_var_extrap.std(dim='date')*tstar_ext/LWmerge_ext.time.std(dim='date')/sqrt(neff_ext-2)).values
conf_int_ext_unobs=(LWmerge_ext.MOCdepth_var.std(dim='date')*tstar_ext/LWmerge_ext.time.std(dim='date')/sqrt(neff_ext-2)).values
```

```python
conf_int_ek=(lw_ek.std(dim='date')*tstar/LWmerge.time.std(dim='date')/sqrt(neff-2)).values
conf_int_ek_ext=(lw_ek.std(dim='date')*tstar_ext/LWmerge_ext.time.std(dim='date')/sqrt(neff_ext-2)).values
```

#### Convert into sigma2 coordinates

Examine differences with depth-space AMOC

```python
denstep=0.05
denvec=arange(30,39,denstep)
denvec_i=denvec[:-1]+denstep/2
```

```python
lev_outer =hstack((0,range(5,5600,10)))
LWmerge=LWmerge.assign_coords(lev_outer=lev_outer)
gridmerge = Grid(LWmerge,periodic=False,coords={'Z':{'center':'depth','outer':'lev_outer'}})
```

```python
trans_sig2=gridmerge.transform(LWmerge.trans_tot_extrap.load(),'Z',denvec,target_data=LWmerge.sigma2_i.fillna(37.5).where(LWmerge.depth<-LWmerge.bathy_i).load(),method='conservative')
```

```python
LWmerge['psi']=LWmerge.trans_tot.sum(dim='dist').cumsum(dim='depth')-LWmerge.trans_tot.sum(dim=['dist','depth'])
```

```python
dmin=0
dmax=2000

LWmerge

LWmerge['MOCdepth_var']=LWmerge.psi.sel(depth=slice(dmin,dmax)).max(dim='depth')

LWmerge['dpthmax']=(['date'],LWmerge.depth.sel(depth=slice(dmin,dmax))[LWmerge.psi.sel(depth=slice(dmin,dmax)).argmax(dim='depth')])
print(LWmerge.dpthmax.mean())

dpick=1000
LWmerge['MOCdepth']=LWmerge.psi.interp(depth=dpick)
LWmerge['MOCdepth_dcomp']=LWmerge.psi.interp(depth=LWmerge.dpthmax.mean())
```

```python
psi_sig2=trans_sig2.sum(dim='dist').cumsum(dim='sigma2_i')-trans_sig2.sum(dim=['dist','sigma2_i'])
```

```python
smin=31
smax=37

LWmerge['MOCsigma_var']=(['date'],psi_sig2.sel(sigma2_i=slice(smin,smax)).max(dim='sigma2_i'))

LWmerge['sigmax']=(['date'],denvec[where((denvec>=smin) & (denvec<smax))][psi_sig2.sel(sigma2_i=slice(smin,smax)).argmax(dim='sigma2_i')])

LWmerge['MOCsigma']=(['date'],psi_sig2.interp(sigma2_i=LWmerge.sigmax.median()))
```

```python
print(LWmerge.MOCdepth_var.mean().values,LWmerge.MOCdepth.mean().values)
print(LWmerge.MOCsigma_var.mean().values,LWmerge.MOCsigma.mean().values)
```
