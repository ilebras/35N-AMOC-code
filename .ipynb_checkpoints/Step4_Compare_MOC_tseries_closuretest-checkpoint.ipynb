```python
for get_acf,tit in zip(tseries_list[:7],tseries_names[:7]):
    acorr = sm.tsa.acf(get_acf, nlags = 20,fft=False)
    ax1=subplot(211)
    ax1.plot(acorr,'o-')
    ax1.axhline(0)
    ax2=subplot(212)
    ax2.plot(cumsum(acorr*hstack([0.5,ones(len(acorr)-2),0.5])),'o-',label=tit)
    ax2.legend(loc=(1.05,0))
    [ax2.axhline(ii) for ii in [1.5]]
```

```python
for get_acf,tit in zip(tseries_list[7:11],tseries_names[7:11]):
    acorr = sm.tsa.acf(get_acf, nlags = 20,fft=False)
    ax1=subplot(211)
    ax1.plot(acorr,'o-')
    ax1.axhline(0)
    ax2=subplot(212)
    ax2.plot(cumsum(acorr*hstack([0.5,ones(len(acorr)-2),0.5])),'o-',label=tit)
    ax2.legend(loc=(1.05,0))
    [ax2.axhline(ii) for ii in [1.75,2.25]]
```

```python
for get_acf,tit in zip(tseries_list[-3:],tseries_names[-3:]):
    acorr = sm.tsa.acf(get_acf, nlags = 20,fft=False)
    ax1=subplot(211)
    ax1.plot(acorr,'o-')
    ax1.axhline(0)
    ax2=subplot(212)
    ax2.plot(cumsum(acorr*hstack([0.5,ones(len(acorr)-2),0.5])),'o-',label=tit)
    ax2.legend(loc=(1.05,0))
    [ax2.axhline(ii) for ii in [1.5,1.75,2.25]]
```

```python
for tseries,tit in zip(tseries_list,tseries_names):# resample with replacement each row
    boot_slopes = []
    n_boots = 10000
    for _ in range(n_boots):
        months=arange(len(tseries))/12
        p=np.polyfit(months,tseries,1)
        tseries_detrend=tseries-(p[0]*months+p[1])
        # sample the rows, same size, with replacement
        sample_df = pd.DataFrame(tseries_detrend.values).sample(n=len(tseries), replace=True)
        # fit a linear regression
        p_fin=polyfit(arange(len(tseries))/12,sample_df,1)
        boot_slopes.append(p_fin[0][0])
    figure()
    hist(boot_slopes)
    title(tit+': '+str(percentile(sort(boot_slopes),2.5))+'/'+str(percentile(sort(boot_slopes),97.5)));
```
